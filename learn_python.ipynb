```python
from collections import defaultdict
```

```python
dict = {'Name':'Ajay', 'Age':23}

defdict = defaultdict(str)
defdict = {'Name':'Ajay', 'Age':23}
```

```python
print(defdict.get('Address'))
print(defdict.get('Address', 'No address given'))

print(dict['Address'])
```

_defaultdict_ can give a default value if a key is not present in dict, else it will return None, but never breaks the code.
