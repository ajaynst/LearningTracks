```python
csv_data["State"] = csv_data["Name of State Road Transport Undertaking (SRTU)"].map(srtu_to_state)
csv_data.drop(["Name of State Road Transport Undertaking (SRTU)"], axis=1, inplace=True) # rm this col
csv_data = csv_data.dropna(subset=["State"], axis=0) # rm NaNs, as total and all is NaN by now

csv_data.head(15)
```

```python
csv_data.columns
```

```python
money_cols = csv_data.columns.drop(['S. No.', 'State'])
money_cols
```

```python
state_level = (csv_data.groupby("State", as_index=False)[money_cols].sum())

state_level
```

```python
# Revenue
state_level['Revenue_sum_16-19'] = (
    state_level['Total Revenue (Rs. Lakh) - 2016-17'] +
    state_level['Total Revenue (Rs. Lakh) - 2017-18'] +
    state_level['Total Revenue (Rs. Lakh) - 2018-19']
)

state_level.drop([
    'Total Revenue (Rs. Lakh) - 2016-17',
    'Total Revenue (Rs. Lakh) - 2017-18',
    'Total Revenue (Rs. Lakh) - 2018-19'
], axis=1, inplace=True)


# Cost
state_level['Cost_sum_16-19'] = (
    state_level['Total Cost (Rs. Lakh) - 2016-17'] +
    state_level['Total Cost (Rs. Lakh) - 2017-18'] +
    state_level['Total Cost (Rs. Lakh) - 2018-19']
)

state_level.drop([
    'Total Cost (Rs. Lakh) - 2016-17',
    'Total Cost (Rs. Lakh) - 2017-18',
    'Total Cost (Rs. Lakh) - 2018-19'
], axis=1, inplace=True)


# Net Profit
state_level['Net_Profit_sum_16-19'] = (
    state_level['Net Profit/Loss (Rs. Lakh) - 2016-17'] +
    state_level['Net Profit/Loss (Rs. Lakh) - 2017-18'] +
    state_level['Net Profit/Loss (Rs. Lakh) - 2018-19']
)

state_level.drop([
    'Net Profit/Loss (Rs. Lakh) - 2016-17',
    'Net Profit/Loss (Rs. Lakh) - 2017-18',
    'Net Profit/Loss (Rs. Lakh) - 2018-19'
], axis=1, inplace=True)


# Profit Before Tax
state_level['Profit_bf_tax_sum_16-19'] = (
    state_level['Profit before Tax (Rs. Lakh) - 2016-17'] +
    state_level['Profit before Tax (Rs. Lakh) - 2017-18'] +
    state_level['Profit before Tax (Rs. Lakh) - 2018-19']
)

state_level.drop([
    'Profit before Tax (Rs. Lakh) - 2016-17',
    'Profit before Tax (Rs. Lakh) - 2017-18',
    'Profit before Tax (Rs. Lakh) - 2018-19'
], axis=1, inplace=True)


state_level.head()
```

Tax Paid = Profit Before Tax − Net Profit (After Tax)

```python
state_level['Tax_paid'] = state_level['Profit_bf_tax_sum_16-19'] - state_level['Net_Profit_sum_16-19']

# ---------------------------------------------------


state_level = state_level.sort_values('Tax_paid')

plt.figure(figsize=(10, 6))
plt.barh(state_level['State'], state_level['Tax_paid'])

plt.xlabel('Tax Paid')
plt.ylabel('State Name')
plt.title('Tax Paid by State')

plt.tight_layout()
plt.show()
```
