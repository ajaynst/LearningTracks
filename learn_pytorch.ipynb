```python
import torch
import torch.nn as nn
```

```python
x = torch.tensor([1.0, 2.0, 3.0])
y = x * 2

print(x, y)
```

```python
x = torch.tensor(3.0, requires_grad=True) # requires_grad track ops on x
y = x ** 2 # autograd records op
y.backward() # computes dy/dx => 2x, where x = 3 ==> 2x3 = 6

print(x.grad) # should be 6, a tensor
```

```python
model = nn.Linear(4, 1)
x = torch.randn(10, 4)
y = model(x)

print(model)
print(x)
print(y)
```

```python
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(4, 8)
        self.fc2 = nn.Linear(8, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        return self.fc2(x)

model = MyModel()
print(model)
```
