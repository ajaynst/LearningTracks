```python
words = ["ab", "bc", "bcd", "cde"]
```

```python
chars = []
char_count_dict = {}

for word in words:
    print(word, end=": ")
    for char in word:
        chars.append(char)
        if char not in char_count_dict:
            char_count_dict[char] = 1
        else:
            char_count_dict[char] += 1    
        print(char, end=" ")
    print()

print("chars: ", list(set(chars))) # rm dups and print chars
print("char_count_dict: ", char_count_dict)
```

```python
chars = []
word_count_dict = {}
char_count_dict = {}

for word in words:
    print(word, end=": ")
    if char not in word_count_dict:
            word_count_dict[word] = 1
    else:
        word_count_dict[word] += 1    
    for char in word:
        chars.append(char)
        if char not in char_count_dict:
            char_count_dict[char] = 1
        else:
            char_count_dict[char] += 1    
        print(char, end=" ")
    print()

print()
print("words: ", words)
print("word_count_dict: ", word_count_dict)
print()
print("chars: ", list(set(chars))) # rm dups and print chars
print("char_count_dict: ", char_count_dict)
```
